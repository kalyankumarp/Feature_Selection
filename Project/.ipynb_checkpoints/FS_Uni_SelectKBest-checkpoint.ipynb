```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest, chi2, f_classif
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('train.csv')
```

```python
data.info()
```

```python
data[['blue', 'dual_sim', 'four_g', 'three_g','touch_screen', 'wifi']] = data[['blue', 'dual_sim', 'four_g','three_g', 'touch_screen', 'wifi']].astype('bool')
```

```python
data.head()
```

```python
data.info()
```

```python
x = data.iloc[:,:20]
y = data.iloc[:,-1:]
```

```python
x.head()
x.columns
```

```python
y.head()
```

```python
newdf = list(x.select_dtypes(include=[np.number]))
newdf
```

```python
scaler = StandardScaler()
x[['battery_power', 'clock_speed',  'fc',  'int_memory',  'm_dep',  'mobile_wt', 
   'n_cores',  'pc',  'px_height', 'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time']]= scaler.fit_transform(x[['battery_power',
 'clock_speed',
 'fc',
 'int_memory',
 'm_dep',
 'mobile_wt',
 'n_cores',
 'pc',
 'px_height',
 'px_width',
 'ram',
 'sc_h',
 'sc_w',
 'talk_time']])
```

```python
x.head()
```

```python
x[['battery_power', 'clock_speed',  'fc',  'int_memory',  'm_dep',  'mobile_wt', 
   'n_cores',  'pc',  'px_height', 'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time']] = x[['battery_power', 'clock_speed',  'fc',  'int_memory',  'm_dep',  'mobile_wt', 
   'n_cores',  'pc',  'px_height', 'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time']] +10
```

```python
x.head()
```

```python
features = x.columns
features
```

```python
bestfeatures = SelectKBest(score_func = f_classif, k=10)
print(bestfeatures)
```

```python
fit = bestfeatures.fit(x,y)
```

```python
fit.get_support()
```

```python
col = fit.get_support()
new_features = []
for bool, feature in zip(col, features):
    if bool:
        new_features.append(feature)

new_features
```

```python
data1 = pd.DataFrame(x, columns=new_features)
```

```python
data1.info()
```

```python
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(x.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']

print(featureScores.nlargest(10,'Score'))

```

```python
bestcols = featureScores.nlargest(10,'Score')['Specs']
bestcols
```

```python
datab = x[bestcols]
```
