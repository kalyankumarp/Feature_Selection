```python
x[['battery_power', 'clock_speed',  'fc',  'int_memory',  'm_dep',  'mobile_wt', 
   'n_cores',  'pc',  'px_height', 'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time']] = x[['battery_power', 'clock_speed',  'fc',  'int_memory',  'm_dep',  'mobile_wt', 
   'n_cores',  'pc',  'px_height', 'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time']] +10
```

```python
features = x.columns
features
```

```python
model = ExtraTreesClassifier()
model.fit(x,y)

print(model.feature_importances_)

```

```python
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=x.columns)
feat_importances.nlargest(10).plot(kind='barh')
```

```python
features = x.columns
new_features = []
for t, feature in zip(feat_importances, features):
    if t > feat_importances.median():
        new_features.append(feature)

new_features
```

```python
data1 = pd.DataFrame(x, columns=new_features)
```
